```python
import numpy as np
import scipy as sci
from scipy.fftpack import fft
from matplotlib.axis import Axis
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import pandas as pd
from numpy.fft import rfft
import sys
np.set_printoptions(threshold=sys.maxsize) #Shows all values in array.

CSVDataFarBlue = pd.read_csv("FarBlueDetune.csv")
CSVDataBlue = pd.read_csv("BlueDetune.csv")
CSVDataRed = pd.read_csv("RedDetuning.csv")
CSVDataFarRed = pd.read_csv("FarRedDetune.csv")

FBx1 = CSVDataFarBlue['X1']
FBy1 = CSVDataFarBlue['Y1']
FBx2 = CSVDataFarBlue['X2']
FBy2 = CSVDataFarBlue['Y2']

Bx1 = CSVDataBlue['X1']
By1 = CSVDataBlue['Y1']
Bx2 = CSVDataBlue['X2']
By2 = CSVDataBlue['Y2']
Bx3 = CSVDataBlue['X3']
By3 = CSVDataBlue['Y3']
Bx4 = CSVDataBlue['X4']
By4 = CSVDataBlue['Y4']

Rx1 = CSVDataRed['X1']
Ry1 = CSVDataRed['Y1']
Rx2 = CSVDataRed['X2']
Ry2 = CSVDataRed['Y2']
Rx3 = CSVDataRed['X3']
Ry3 = CSVDataRed['Y3']
Rx4 = CSVDataRed['X4']
Ry4 = CSVDataRed['Y4']

FRx1 = CSVDataFarRed['X1']
FRy1 = CSVDataFarRed['Y1']
FRx2 = CSVDataFarRed['X2']
FRy2 = CSVDataFarRed['Y2']

RECORDING_RATE = 1*10**-6 #Input recording rate for uncertainty error prop later. Usually either 1 | 2 micro seconds.
```

```python
def plotBurst(): #Choose to plot a graph of choice.
    plt.plot(FRx2[0:70000],FRy2[0:70000])
    plt.grid(True)
    plt.title("Non-offsetted waveform")
    plt.xlabel("Time (ms)")
    plt.ylabel("Voltage (mV)")
plotBurst()
```

```python
#DECAYS -"D#"
CSVDataD1 = By1[0:70000] = By2[0:70000].astype(float) #Setting String Dataframe -> float.
FourierD1 = sci.fft(CSVDataD1) #Fourier Complex of Y values same data set as excel. (1D FFT) - should do DFT also.(although 1/2 values)
YMagD1 = abs(FourierD1)
tD1 = np.linspace(0,69999,70000) #Samples of 0 -> 28672 for DFT FREQ.
TD1 = (1/(RECORDING_RATE))/(70000) #S/N (Sample rate / total points.)
FourierFREQD1 = TD1*tD1 #FFT / DFT FREQ.

CSVDataD2 = By2[0:70000] = By2[0:70000].astype(float) #Setting String Dataframe -> float.
FourierD2 = sci.fft(CSVDataD2) #Fourier Complex of Y values same data set as excel. (1D FFT) - should do DFT also.(although 1/2 values)
YMagD2 = abs(FourierD2)
tD2 = np.linspace(0,69999,70000) #Samples of 0 -> 28672 for DFT FREQ.
TD2 = (1/(RECORDING_RATE))/(70000) #S/N (Sample rate / total points.)
FourierFREQD2 = TD2*tD2 #FFT / DFT FREQ.

CSVDataD3 = By3[0:70000] = By3[0:70000].astype(float) #Setting String Dataframe -> float.
FourierD3 = sci.fft(CSVDataD3) #Fourier Complex of Y values same data set as excel. (1D FFT) - should do DFT also.(although 1/2 values)
YMagD3 = abs(FourierD3)
tD3 = np.linspace(0,69999,70000) #Samples of 0 -> 28000 for DFT FREQ.
TD3 = (1/(RECORDING_RATE))/(70000) #S/N (Sample rate / total points.)
FourierFREQD3 = TD3*tD3 #FFT / DFT FREQ.

CSVDataD4 = By4[0:70000] = By4[0:70000].astype(float) #Setting String Dataframe -> float.
FourierD4 = sci.fft(CSVDataD4) #Fourier Complex of Y values same data set as excel. (1D FFT) - should do DFT also.(although 1/2 values)
YMagD4 = abs(FourierD4)
tD4 = np.linspace(0,69999,70000) #Samples of 0 -> 28000 for DFT FREQ.
TD4 = (1/(RECORDING_RATE))/(70000) #S/N (Sample rate / total points.)
FourierFREQD4 = TD4*tD4 #FFT / DFT FREQ.

#TD"#" = 1/1*10^-6... as this iteration of data recording was every 1microS where as before was 2microS 
```

```python
#DECAY GRAPHING
#DONT FORGET: DFFT|FFT Mirrors values at halfway point, so 50000-> is a mirror. Thats the reason for limit on smaller graph.
def DecayD1():#Zoom-in on the peak ~ 1382 to show datapoints more accurately.
    x = FourierFREQD1 #Replotting with zoom feature.
    y = YMagD1

    ax = plt.subplot(111) #Layout of of main graph. (X / Y dimensions)
    ax.plot(x, y)
    plt.title("All Detune Decays")
    plt.xlabel("FFT FREQ (Hz)")
    plt.ylabel("Magnitude")
    ax.set_ylim(0, 30000) #Y limit for main graph.
    ax.set_xlim(3100, 4000)
    #plt.grid(True)

    ax0 = plt.axes([0.2, 1.00, .6, .3])
    ax0.plot(x, y)
    ax0.set_xlim(0,25000)
    
def DecayD2():#Zoom-in on the peak ~ 1382 to show datapoints more accurately.
    x = FourierFREQD2 #Replotting with zoom feature.
    y = YMagD2

    ax = plt.subplot(111) #Layout of of main graph. (X / Y dimensions)
    ax.plot(x, y)
    ax.set_ylim(0, 30000) #Y limit for main graph.
    ax.set_xlim(3100, 4000)
    #plt.grid(True)

    ax1 = plt.axes([0.2, 1.00, .6, .3])
    ax1.plot(x, y)
    ax1.set_xlim(0,25000)
    
def DecayD3():#Zoom-in on the peak ~ 1382 to show datapoints more accurately.
    x = FourierFREQD3 #Replotting with zoom feature.
    y = YMagD3

    ax = plt.subplot(111) #Layout of of main graph. (X / Y dimensions)
    ax.plot(x, y)
    ax.set_ylim(0, 30000) #Y limit for main graph.
    ax.set_xlim(3100, 4000)
    #plt.grid(True)

    ax1 = plt.axes([0.2, 1.00, .6, .3])
    ax1.plot(x, y)
    ax1.set_xlim(0,25000)    

def DecayD4():#Zoom-in on the peak ~ 1382 to show datapoints more accurately.
    x = FourierFREQD4 #Replotting with zoom feature.
    y = YMagD4

    ax = plt.subplot(111) #Layout of of main graph. (X / Y dimensions)
    ax.plot(x, y)
    ax.set_ylim(0, 30000) #Y limit for main graph.
    ax.set_xlim(3100, 4000)
    #plt.grid(True)

    ax1 = plt.axes([0.2, 1.00, .6, .3])
    ax1.plot(x, y)
    ax1.set_xlim(0,25000)
DecayD1()
DecayD2()
DecayD3()
DecayD4()
```
